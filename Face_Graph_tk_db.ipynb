```python
from tkinter import *
import requests
import json
import sqlite3
# import xlwt
conn=sqlite3.connect("/home/sandip/code/Python/Web_Scrapping/details.db")
cur=conn.cursor()

root = Tk()
root.minsize(300,300) 
root.title("Facebook Profile Data")
token=""
user_id=""
home_town_text=""
name_text=""
friends_count=""
birthday_text=""
gender_text=""
email_text=""
location=""
#urls
def feature_extract():
    name='https://graph.facebook.com/'+user_id+'?fields=name&access_token='+token
    home_town='https://graph.facebook.com/'+user_id+'?fields=hometown&access_token='+token
    friends='https://graph.facebook.com/'+user_id+'?fields=friends&access_token='+token
    birthday="https://graph.facebook.com/v4.0/"+user_id+"?fields=birthday&access_token="+token
    gender="https://graph.facebook.com/v4.0/"+user_id+"?fields=gender&access_token="+token
    email="https://graph.facebook.com/v4.0/"+user_id+"?fields=email&access_token="+token
    location="https://graph.facebook.com/v4.0/"+user_id+"?fields=location&access_token="+token


    home_town_json=requests.get(home_town)
    name_json=requests.get(name)
    friends_json=requests.get(friends)
    birthday_json=requests.get(birthday)
    gender_json=requests.get(gender)
    email_json=requests.get(email)


    home_town_dict=home_town_json.json()
    name_dict=name_json.json()
    friends_dict=friends_json.json()
    birthday_dict=birthday_json.json()
    gender_dict=gender_json.json()
    email_dict=email_json.json()
    
    global home_town_text
    global name_text
    global friends_count
    global birthday_text
    global email_text
    global gender_text
    
    home_town_text=home_town_dict['hometown']['name']
    name_text=name_dict['name']
    friends_count=friends_dict['friends']['summary']['total_count']
    birthday_text=birthday_dict['birthday']
    gender_text=gender_dict['gender']
    email_text=email_dict['email']

```

```python
f1=Label(root, text ="Enter Access Token")
e1 =Entry(root)
e1.config(bg="light blue")
f2 =Label(root, text ="Enter User ID")
e2 =Entry(root)
e2.config(bg="light blue")
l =  Label(root)
Answer =  Label(root)

def callback():
    global token,user_id
    token=e1.get()
    user_id=e2.get()
    feature_extract()
    global home_town_text
    global name_text
    global friends_count
    global birthday_text
    global email_text
    global gender_text
    l.config(text="Name: {}\nEmail: {}\nGender: {}\nFriends: {}\nBirthday: {}\nHomeTown: {}".format(name_text,email_text,gender_text,friends_count,birthday_text,home_town_text))


def store():
    global home_town_text
    global name_text
    global friends_count
    global birthday_text
    global email_text
    global gender_text
    
#     head=["home_town_text","name_text","friends_count","birthday_text","email_text","gender_text"]
#     details=[home_town_text,name_text,friends_count,birthday_text,email_text,gender_text]
    cur.execute("drop table  IF EXISTS facedata")
    cur.execute("create table facedata (home_town text,name text,friends INT,birthday text,email text,gender text)")
    cur.execute('''insert into facedata (home_town,name,friends,birthday,email,gender) VALUES(?,?,?,?,?,?)''',(home_town_text,name_text,friends_count,birthday_text,email_text,gender_text))
    conn.commit()
    Answer.config(text="\nIt's Done")
    
def destroy():
    root.destroy()

    
b = Button(root, text="OK", command=callback)

s = Label(root, text ="\nwant to store??")
Yes = Button(root, text="Yes", command=store)
No = Button(root, text="No/Exit", command=destroy)


for widget in (f1,e1,f2,e2,b,l,s,Yes,No,Answer):
    widget.pack()
b.mainloop()
```
